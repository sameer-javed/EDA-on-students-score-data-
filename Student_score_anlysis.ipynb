```python
df['WklyStudyHours'] = df['WklyStudyHours'].str.replace("10-May","5-10")
df.head(10)
```

```python
# Filling Null value's in the siblings column by their mean 
sib_mean = df["NrSiblings"].mean()
df["NrSiblings"] = df['NrSiblings'].fillna(sib_mean).astype(int)
```

```python
df.head(10)
```

```python
# Gender desrtribution 
data_label = sns.countplot(data=df , x = "Gender")
data_label.bar_label(data_label.containers[0]) # Adds data labels into the chart
```

From the Above chart we have analyzed that Number of females present in the data are more than number of males 

```python
# it shows how  parents education affects childrens  study score
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb ,annot=True)
plt.show()
```

from the above chart we have concluded that parents education is impacting childrens education score

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb1
```

```python
plt.figure(figsize=(4,4))
plt.title('Study score destribution by parents marental status')
sns.heatmap(gb1 ,annot=True)
plt.show()
```

from the above chart we concluded that the parents marental status is not affecting childrens study score

```python
gb2 = df.groupby("NrSiblings").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb2
```

```python
plt.figure(figsize=(4,4))
plt.title('Study score destribution by siblings')
sns.heatmap(gb2 ,annot=True)
plt.show()
```

According to the above chart Number of siblings are not affecting students study score

## Detribution by Ethnic Group

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

mylst = [groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]
l = ['groupA', 'groupB', 'groupC', 'groupD','groupE']
plt.title('destribution by Ethnic group')
plt.pie(mylst, labels = l , autopct="%1.2f%%")
plt.show()
```
